```python
import pandas
dataframe = pandas.read_csv('studentscores.csv')
dataframe
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import classification_report
```

```python
# Extract features and target
X = dataframe['Hours'].to_numpy().reshape(-1, 1)
y = dataframe['Scores'].to_numpy()
```

```python
# Split the data into training and testing sets (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Create a Linear Regression model
model = LinearRegression()

# Fit the model on training data
model.fit(X_train, y_train)
```

```python
# Plot the training and testing data
plt.figure(figsize=(10, 6))
plt.scatter(X_train, y_train, color="yellow", label="Training Data")
plt.scatter(X_test, y_test, color="red", label="Testing Data")
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.title("Training and Testing Data")
plt.legend()
plt.show()
```

```python
# Make predictions
y_pred = model.predict(X_test)
print("Predictions:", y_pred)
print("Actual values:", y_test)
```

```python
# Plot predicted line
plt.scatter(X_test, y_test, color='blue', label='Actual values')
plt.plot(X_test, y_pred, color='red', label='Predicted line')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title('Linear Regression - Predictions vs Actuals')
plt.legend()
plt.show()
```

```python
# Calculate regression metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
```

```python
# Print regression performance metrics
print("Regression Performance Metrics:")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R² Score: {r2:.2f}")
```

```python
# For confusion matrix, convert the continuous predictions to binary (threshold example)
# We are defining a threshold value to convert continuous predictions to binary classification
threshold = 50  # You can change this value based on your domain knowledge or data
y_test_binary = (y_test > threshold).astype(int)  # True if greater than threshold, else False
y_pred_binary = (y_pred > threshold).astype(int)  # Convert predictions similarly
```

```python
# Now compute confusion matrix and display
cm = confusion_matrix(y_test_binary, y_pred_binary)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Below Threshold", "Above Threshold"])
```

```python
# Plot confusion matrix
disp.plot(cmap="Blues")
plt.title("Confusion Matrix - Binary Classification from Regression")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
# Print classification report for further details
print("Classification Report:")
print(classification_report(y_test_binary, y_pred_binary))
```
